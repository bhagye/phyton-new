```python
import matplotlib.pyplot as pt 
import numpy as np 

angle = np.linspace( 0 , 2 * np.pi  , 150)

radius = 0.4

x = radius * np.cos( angle )
y = radius * np.sin( angle )

figure, axes = plt.subplots( 1 )

axes.plot( x, y )
axes.set_aspect( 1 )

plt.title( 'parametric equation circle' )
plt.show()
```
